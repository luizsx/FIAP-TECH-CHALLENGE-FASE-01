Se o NPS for usado de forma inadequada, pode levar a interpretações equivocadas.<br>

<b>0 a 6:</b> Detratores<br>
Clientes insatisfeitos, com maior chance de reclamar ou não voltar.<br>
<b>7 e 8:</b> Neutros (ou Passivos)<br>
Clientes ok, mas não são leais nem promovem a marca.<br>
<b>9 e 10:</b> Promotores<br>
Clientes satisfeitos, com alta chance de recomendar.

```python
#criacao classificacao NPS
condicoes = [
    dado_origem['nps_score'] <= 6,
    dado_origem['nps_score'] <= 8
]

valores = ['Detratores', 'Neutros']

dado_origem['classificacao'] = np.select(condicoes, valores, default='Promotores')

```

```python
#Distribuição do score de NPS
sns.histplot(
    data=dado_origem, 
    x='nps_score', 
    kde=True,
    bins=20,
    alpha=0.7
)

plt.axvline(dado_origem['nps_score'].mean(), linestyle='--')

plt.title('Distribuição do NPS')
plt.xlabel('Score de NPS')
plt.ylabel('Frequência')
#plt.grid(axis='y', linestyle='--', alpha=0.3)

plt.show()
```

```python
dado_origem['classificacao'].value_counts()
```

<b>C) Existe algum “ponto de ruptura” na experiência do cliente?</b>

<b>RESPOSTA:</b><br>

```python
dado_origem.groupby('delivery_delay_days')['nps_score'].mean()
```

```python
sns.lineplot(
    data=dado_origem,
    x='delivery_delay_days',
    y='nps_score'
)
```

```python
bins = [0, 2, 5, 10]
labels = ['0-2', '3-5', '6+']

dado_origem['faixa_atraso'] = pd.cut(
    dado_origem['delivery_delay_days'],
    bins=bins,
    labels=labels
)

dado_origem.groupby('faixa_atraso')['nps_score'].mean()
```

- Observa-se um ponto de ruptura claro na experiência do cliente a partir de aproximadamente 3 dias de atraso na entrega. 
- Enquanto atrasos de até 2 dias ainda mantêm um NPS relativamente estável, a partir desse ponto há uma queda acentuada na satisfação, indicando que atrasos mais longos impactam significativamente a percepção do cliente.

<b>D) Que tipo de cliente tende a ter NPS mais alto ou mais baixo? <br>
- Imagine que você está explicando isso para um(a) gerente de operações que não entende estatística.

<b>RESPOSTA:</b><br>
- Clientes que têm uma experiência mais fluida tendem a dar notas mais altas. Ou seja, quando o pedido chega no prazo, sem problemas, e não precisa acionar o atendimento, o cliente fica satisfeito e vira promotor.

- Por outro lado, clientes que enfrentam algum tipo de problema ao longo da jornada são os que mais dão notas baixas.<br>
 Principalmente quando há atraso na entrega, necessidade de abrir reclamações ou demora para resolver um problema. Quanto mais esses pontos se acumulam, pior é a percepção do cliente.

- Outro ponto importante é que existe um limite de tolerância: pequenos atrasos ainda são aceitos, mas quando passam de alguns dias, a satisfação cai bastante.<br>

Resumo:<br>
NPS alto: entrega no prazo, sem problemas, experiência simples<br>
NPS baixo: atraso, reclamação e demora na resolução<br><br>

O cliente até tolera pequenos erros, mas não tolera fricção acumulada na experiência”
